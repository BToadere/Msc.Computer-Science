```sage


def flow_augmenting_step(V, A, c, f, s, t):
    """
    Implementación de un paso del algoritmo de aumento de flujo.
    V: lista de vértices
    A: lista de arcos (tuplas (u,v))
    c: diccionario de capacidades, c[(u,v)] = capacidad
    f: diccionario de flujos, f[(u,v)] = flujo actual
    s: vértice fuente
    t: vértice sumidero
    """
    # Construir el grafo residual Df
    # Para cada arco (u,v): 
    #   - si f(u,v) < c(u,v), entonces (u,v) está en Af con capacidad residual c(u,v)-f(u,v)
    #   - si f(u,v) > 0, entonces (v,u) está en Af con capacidad residual f(u,v)
    Af = {}
    for (u,v) in A:
        if f[(u,v)] < c[(u,v)]:
            Af[(u,v)] = c[(u,v)] - f[(u,v)]
        if f[(u,v)] > 0:
            Af[(v,u)] = f[(u,v)]

    # Buscar un camino aumentante de s a t en el grafo residual
    from collections import deque
    padres = {x: None for x in V}
    capacidades_camino = {x: 0 for x in V}
    capacidades_camino[s] = float('inf')
    
    cola = deque([s])
    while cola:
        u = cola.popleft()
        for v in V:
            if (u,v) in Af and Af[(u,v)] > 0 and padres[v] is None and v != s:
                padres[v] = u
                capacidades_camino[v] = min(capacidades_camino[u], Af[(u,v)])
                if v == t:
                    # Se encontró un camino aumentante
                    # Reconstruir el camino y aumentar el flujo
                    camino = []
                    alpha = float('inf')
                    x = t
                    while x != s:
                        camino.append((padres[x], x))
                        if (padres[x], x) in Af:
                            alpha = min(alpha, Af[(padres[x], x)])
                        x = padres[x]
                    camino.reverse()
                    
                    # Aumentar el flujo en el camino
                    for (p,q) in camino:
                        if (p,q) in A: 
                            f[(p,q)] += alpha
                        else: 
                            # (p,q) no es arco original, debe ser residual inverso
                            # por lo tanto (q,p) es el arco original y disminuimos su flujo
                            f[(q,p)] -= alpha
                    return f, None
                cola.append(v)

    # Si no hay camino aumentante, encontrar el corte
    # U = {u ∈ V | existe camino en Df desde s a u}
    # Hemos marcado con padres los alcanzables desde s
    U = {u for u in V if padres[u] is not None or u == s}
    # δout(U) son los arcos que salen de U a V\U
    delta_out = [(u,v) for (u,v) in A if u in U and v not in U]
    return None, delta_out
# Crear un grafo dirigido con pesos
D = DiGraph(weighted=True)

# Agregar nodos
D.add_vertices(['s', 'a', 'b', 'c', 'd', 'e', 'f', 't'])

# Agregar aristas con pesos
D.add_edge('s', 'a', 3)  # Arista de 's' a 'a' con peso 5
D.add_edge('s', 'b', 3)
D.add_edge('s', 'c', 4)
D.add_edge('a', 'e', 3)
D.add_edge('b', 'd', 3)
D.add_edge('c', 'f', 4)
D.add_edge('d', 't', 3)
D.add_edge('e', 't', 3)
D.add_edge('f', 't', 4)

# Ejemplo de uso:
# Suponiendo un pequeño ejemplo de grafo
V = D.vertices()                # Lista de vértices
A = [(u,v) for (u,v,w) in D.edges()]   # Lista de arcos
c = {(u,v): w for (u,v,w) in D.edges()}  # Capacidades

# Ejecución de un paso de aumento
nuevo_f, corte = flow_augmenting_step(V, A, c, f, 's', 't')

if nuevo_f is not None:
    print("Se encontró un camino aumentante. Nuevo flujo:")
    print(nuevo_f)
else:
    print("No hay camino aumentante. Corte encontrado:")
    print(corte)
```

```sage
from collections import deque

def build_residual_graph(A, c, f):
    """
    Construye el grafo residual en formato de lista de adyacencia.
    """
    Af = {}
    for (u, v) in A:
        if f[(u, v)] < c[(u, v)]:  # Arco forward
            Af.setdefault(u, []).append(v)
        if f[(u, v)] > 0:  # Arco backward
            Af.setdefault(v, []).append(u)
    return Af

def find_path(Af, s, t):
    """
    Encuentra un camino de s a t en el grafo residual Af usando BFS.
    """
    visited = set()
    queue = deque([(s, [s])])  # Tupla con nodo actual y camino acumulado
    
    while queue:
        current, path = queue.popleft()
        if current == t:
            return path  # Camino encontrado
        for neighbor in Af.get(current, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None  # No se encontró camino

def augmenting_flow_step(V, A, c, f, s, t, verbose=True):
    """
    Realiza un paso del algoritmo de Ford-Fulkerson para aumentar el flujo.
    """
    # Construir el grafo residual
    Af = build_residual_graph(A, c, f)
    if verbose:
        print("Grafo residual Af:", Af)

    # Buscar un camino aumentante
    path = find_path(Af, s, t)

    if path is not None:
        if verbose:
            print("Camino aumentante encontrado:", path)

        # Determinar el flujo máximo que puede aumentar (α)
        alphas = []
        for i in range(len(path) - 1):
            u, v = path[i], path[i + 1]
            if (u, v) in A:  # Arco forward
                alphas.append(c.get((u, v), 0) - f.get((u, v), 0))
            elif (v, u) in A:  # Arco backward
                alphas.append(f.get((v, u), 0))
        
        alpha = min(alphas)
        if verbose:
            print("Valores de capacidad residual (σ):", alphas)
            print("Flujo aumentante α:", alpha)

        # Actualizar el flujo
        for i in range(len(path) - 1):
            u, v = path[i], path[i + 1]
            if (u, v) in A:  # Arco forward
                f[(u, v)] += alpha
            elif (v, u) in A:  # Arco backward
                f[(v, u)] -= alpha

        # Calcular el nuevo valor del flujo
        new_flow_value = sum(f[(s, v)] for v in V if (s, v) in A)
        if verbose:
            print("Capacidades de los arcos:        ", c)
            print("Flujo actualizado en los arcos:  ", f)
            print("Nuevo valor del flujo máximo:", new_flow_value)
        return f

    else:
        if verbose:
            print("No existe camino aumentante s-t.")

        # Determinar los nodos alcanzables desde s
        U = set()
        queue = deque([s])
        U.add(s)
        while queue:
            u = queue.popleft()
            for v in Af.get(u, []):
                if v not in U:
                    U.add(v)
                    queue.append(v)

        # Determinar el corte mínimo
        cut = [(u, v) for (u, v) in A if u in U and v not in U]
        val_f = sum(f[(s, v)] for v in V if (s, v) in A)
        capacity_cut = sum(c[(u, v)] for (u, v) in cut)

        if verbose:
            print("Corte mínimo δ^{out}(U):", cut)
            print("Capacidad del corte c(δ^{out}(U)):", capacity_cut)
            print("Valor del flujo máximo value(f):", val_f)

        return cut

def maximum_flow_algorithm(V, A, c, s, t, verbose=True):
    """
    Algoritmo de flujo máximo utilizando augmenting_flow_step.
    
    Parámetros:
        V: Lista de nodos.
        A: Lista de arcos (u, v).
        c: Diccionario de capacidades de los arcos, c[(u, v)].
        s: Nodo fuente.
        t: Nodo sumidero.
        verbose: Si True, imprime detalles en cada iteración.
    
    Retorna:
        - El flujo máximo `f` como un diccionario.
        - El corte mínimo `cut` como una lista de arcos.
        - El valor del flujo máximo `value_f`.
    """
    # Inicializar flujo nulo (f0)
    f = {a: 0 for a in A}

    iteration = 0
    while True:
        if verbose:
            print(f"\nIteración {iteration}:")
        # Intentar realizar un paso del flujo aumentante
        result = augmenting_flow_step(V, A, c, f, s, t, verbose=verbose)

        if isinstance(result, list):  # Se devolvió un corte mínimo (δout(U))
            cut = result
            break
        else:  # Se devolvió un flujo actualizado
            f = result
        iteration += 1

    # Calcular el valor del flujo máximo
    value_f = sum(f[(s, v)] for v in V if (s, v) in A)

    if verbose:
        print("\n=== Resultado final ===")
        print("Flujo máximo:", f)
        print("Corte mínimo δ^{out}(U):", cut)
        print("Valor del flujo máximo:", value_f)

    return f, cut, value_f

# Datos de ejemplo
V = ['s', 'a', 'b', 'c', 'd', 'e', 'f', 't']
c = {
    ('s', 'a'): 3, ('s', 'b'): 3, ('s', 'c'): 4,
    ('b', 'd'): 3, ('a', 'e'): 3, ('c', 'f'): 4,
    ('d', 't'): 3, ('e', 't'): 3, ('f', 't'): 4
}
A = list(c.keys())
s = 's'
t = 't'

# Ejecutar el algoritmo
f, cut, value_f = maximum_flow_algorithm(V, A, c, s, t, verbose=False)

print(f, cut,  value_f, sep='\n')
```
